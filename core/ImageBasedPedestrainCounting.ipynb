```python
from imageai.Detection import ObjectDetection
import os
import cv2
import pandas as pd
import random
import numpy as np

algos = {
        "resnet": "resnet50_coco_best_v2.0.1.h5",
        "yolov3": "yolo.h5",
        "yolo_tiny": "yolo-tiny.h5"
    } 

speed = ["normal", "fast", "faster" , "fastest", "flash"]

###########################
execution_path = os.getcwd()
dataset_path = os.path.join(execution_path,"dublin_night_baseline")
if not os.path.isdir('.\\algos_dublin_night'):
        os.makedirs('.\\algos_dublin_night')
dir_path = '.\\algos_dublin_night'
#######################################
def put_text_to_img(img,text,pos=(50, 50),fontColor=(0, 0, 255),lineType=2):
        if img is None:
            print("Put text to a none image.")
            return

        font = cv2.FONT_HERSHEY_SIMPLEX
        fontScale = 1

        cv2.putText(img, text, pos, font, fontScale, fontColor, lineType)
####################
        
def stroe_info_in_df_csv(image_prefix="counting_person"):
        df = pd.DataFrame(
            np.array(res), columns=['image_name','algo', 'detected_num'])
        df["counted_num"] = ""  #only for baseline
        df.to_csv(
            path_or_buf=os.path.join(dir_path, "%s.csv" %
                                     image_prefix))
        return df

##################################################################################
def detector_(algo, image,speed="normal"):
        if algo == "resnet":
            detector.setModelTypeAsRetinaNet()
            detector.setModelPath(
                os.path.join(execution_path, algos["resnet"]))
        elif algo == "yolov3":
            detector.setModelTypeAsYOLOv3()
            detector.setModelPath(
                os.path.join(execution_path, algos["yolov3"]))
        elif algo == "yolo_tiny":
            detector.setModelTypeAsTinyYOLOv3()
            detector.setModelPath(
                os.path.join(execution_path, algos["yolo_tiny"]))

        detector.loadModel(detection_speed=speed)
        custom_objects = detector.CustomObjects(person=True)
        detections = detector.detectCustomObjectsFromImage(custom_objects=custom_objects, input_image=os.path.join(dataset_path,image), output_image_path=os.path.join(dir_path , "%s_%s"%(algo,image)), minimum_percentage_probability=30)
        img = cv2.imread(os.path.join(dir_path, "%s_%s"%(algo,image)))
        put_text_to_img(img, "The number of person:%s " % str(len(detections)))
        cv2.imwrite(os.path.join(dir_path, "%s_%s"%(algo,image)), img)
        return "%s_%s"%(algo,image), algo, len(detections)
########################################
```

```python
execution_path = os.getcwd()
dataset_path = os.path.join(execution_path,"dublin_night_baseline")
if not os.path.isdir('.\\algos_dublin_night'):
        os.makedirs('.\\algos_dublin_night')
dir_path = '.\\algos_dublin_night'

files= os.listdir(dataset_path)
res = []
images = []
for i in range(15):
#     random.shuffle(files)
    index = random.randrange(0, len(files))
    if files[index] not in images: 
        images.append(files[index])
    else: 
        print("dup")
    
print(images, len(images))
    
for algo in algos:
    detector = ObjectDetection()
    print('----------------------------')
    for im in images: 
        frame_res = detector_(algo,im)
        res.append(frame_res)
        print(algo,im)
stroe_info_in_df_csv(image_prefix="Algos_dublin_night")

    
```
