```python
from imageai.Detection import ObjectDetection
import os
from time import time
from selenium import webdriver
import cv2
import streamlink

execution_path = os.getcwd()
detector = ObjectDetection()

timesquare = 'https://www.earthcam.com/usa/newyork/timessquare/?cam=tsrobo1'
dublin = 'https://www.earthcam.com/world/ireland/dublin/?cam=templebar'

def resnet():
    detector.setModelTypeAsRetinaNet()
    detector.setModelPath( os.path.join(execution_path , "resnet50_coco_best_v2.0.1.h5"))
    print("the current algo is resnet")
def yolo():
    detector.setModelTypeAsYOLOv3()
    detector.setModelPath( os.path.join(execution_path , "yolo.h5"))
    print("the current algo is yolov3")
def yolo_tiny():
    detector.setModelTypeAsTinyYOLOv3()
    detector.setModelPath( os.path.join(execution_path , "yolo-tiny.h5"))
    print("the current algo is yolo tiny")

def frame_person_counting(url,imagename="frame",algo="resnet",speed="normal",mpp=30,nim=6,tin=10):
    
    import time
    
    if algo == "resnet":
        resnet()
    elif algo == "yolo":
        yolo()
    elif algo == "yolo_tiny":
        yolo_tiny()

    detector.loadModel(detection_speed=speed)


    count = 0
    custom_objects = detector.CustomObjects(person=True)
    for i in range(nim):
        print('taking frame....',i)
        frame = take_frame(url)      
        detections = detector.detectCustomObjectsFromImage(custom_objects=custom_objects,input_type="array",input_image=frame, output_image_path=os.path.join(execution_path , "%s_%d_result.png" %(imagename,i)), minimum_percentage_probability=mpp)
        
        for eachObject in detections:
            count = count + 1
        print ("the number of person is ", count)
        count = 0
        time.sleep(tin)
    
def take_frame(url):

    streams = streamlink.streams(url)
    q = list(streams.keys())[0]
    stream = streams['%s'%q]
    url = stream.url
    cap = cv2.VideoCapture(url)
    ret, frame = cap.read()
    
    return frame



def screenshot_person_counting(url,imagename="screenshot",algo="resnet",speed="normal",mpp=30,nim=6,tin=10):#nim: number of image, tin: time interval
    
    take_screenshot(url,imagename,nim,tin)

    
    if algo == "resnet":
        resnet()
    elif algo == "yolo":
        yolo()
    elif algo == "yolo_tiny":
        yolo_tiny()

    detector.loadModel(detection_speed=speed)


    count = 0
    custom_objects = detector.CustomObjects(person=True)
    
    for i in range(nim):
        print("--------------------------------")
        print("--------------------------------")
        print("image :  %s_%d.png" %(imagename,i))
        detections = detector.detectCustomObjectsFromImage(custom_objects=custom_objects,input_image=os.path.join(execution_path , "%s_%d.png" %(imagename,i)), output_image_path=os.path.join(execution_path , "%s_%d_result.png" %(imagename,i)), minimum_percentage_probability=mpp)
        for eachObject in detections:
            count = count + 1
        print ("the number of person is ", count)
        count = 0

    
    
def take_screenshot(url,imagename,nim,tin):#nim: number of image, tin: time interval
    import time
    driver = webdriver.Chrome('E:/Googledriver/chromedriver_win32/chromedriver.exe')  # Optional argument, if not specified will search path.
    driver.get(url)
    time.sleep(30)
    for i in range(nim):
        time.sleep(tin) 
        driver.save_screenshot("%s_%d.png" %(imagename,i))
        print('taking image...',i)
    driver.quit()
    
```
